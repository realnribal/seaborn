Let's import the modules 

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

Let's load file `WorldCupMatches.csv`

```python
df = pd.read_csv('WorldCupMatches.csv')
```

Let's inspect DataFrame using `.head()`

```python
#print(df.head())
```

Create a new column in df 

```python
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
print(df.head())
```

#### Set the style of the plot to `whitegrid` 
This will add grid to our plot and make it easier to read

```python
sns.set_style("whitegrid")
```

#### Set the context to `poster` 
Seaborn makes it easy to produce the same plot in **different visual formats.** So you are able to customize the presentation oof your data for the appropriate context. We can do this using `sns.set_context()`. Within his usage, we will use two parameters: `context` and the `font_scale`. You can find more documentation [Here](https://seaborn.pydata.org/generated/seaborn.set_context.html)

```python
sns.set_context("poster",  font_scale=0.8)
```

#### Create __figure__ and __axes__
set the size of the figure to be 12 inches wide and 7 inches tal

```python
f, ax = plt.subplots(figsize=(12,7))

#visualizing how many goals were scored each year the World Cup
ax = sns.barplot(data=df, x='Year', y='Total Goals')


#give a title 
ax.set_title("Average Number Of Goals Scored In World Cup Matches By Year")

plt.show()
```

####  Create a box plot to visualize the distribution of the goals.
**1. Import goals**

```python
df_goals = pd.read_csv('goals.csv')
print(df_goals.head())
```

**Let's change the context**

```python
sns.set_context("notebook",  font_scale=1.25)
```

**Let's Create a second figure**

```python
f, ax2 = plt.subplots(figsize=(12,7))
ax2 = sns.boxplot(data=df_goals, x='year', y='goals', palette='Spectral')
ax2.set_title("Box plot goals over year")
plt.show()
```
